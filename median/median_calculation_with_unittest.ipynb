```python
import unittest

import matplotlib.pylab as plt
import numpy as np
```

```python
def median(data):  
    data.sort()
    
    n = len(data)
    if n % 2 == 0:
        return (data[n//2 - 1] + data[n//2])/2
    else:
        return data[n//2]
```

```python
class TestMedian(unittest.TestCase):

    def test_odd_list(self):
       x = [1,4,3,2]
       assert median(x) == 2.5
        
    def test_even_list(self):
       x = [1,4,3,2,5]
       assert median(x) == 3.0
```

```python
x = [1,6,8,3,4,5,4,7]
median(x)
```

```python
x = [1,6,8,3,4,5,4,7,2]
median(x)
```

```python
#NBVAL_IGNORE_OUTPUT
x = np.arange(100)
y = (np.sin(x / 5) - 0.3)**2

plt.plot(x, y)
plt.axhline(y.mean(), color='0.5')
plt.axhline(median(y), color='0.5', linestyle=':');
```
